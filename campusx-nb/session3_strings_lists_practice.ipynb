A list of strings lines, where each line is a transaction in this format:

YYYY-MM-DD,description,amount,category


amount is a number (can be positive or negative)

negative = expense

positive = income

category is a single word (no commas)

Example line:

2026-01-18,Uber,-240.50,transport


```python

import json

def summarize_transactions(lines: list[str]) -> dict:
    output = {  'balance':0,
                'income_total':0,
                'expense_total':0,
                'category_totals':{},
                'top_expense':None
                }
    income_total = 0
    expense_total = 0
    top_expense = 0
    for line in lines:
        tokens = line.split(',')
        amount = round(float(tokens[2].strip()),2)
        date = tokens[0].strip()
        desc = tokens[1].strip()

        category = tokens[3].strip()

        try:
            output['category_totals'][category] += abs(amount)
        except KeyError:
            output['category_totals'][category] = abs(amount)

        if amount > 0:
            income_total += amount
        else:
            expense_total += amount
            if abs(amount) > abs(top_expense):
                 output['top_expense'] = (date,desc,amount,category)
                 top_expense = amount

        

    output['income_total'] = f"{income_total:.2f}"
    output['expense_total'] = f"{abs(expense_total):.2f}"
    output['balance'] = f"{income_total + expense_total:.2f}"
   
    return output

```

```python
lines = [
  "2026-01-18,Uber,-240.50,transport",
  "2026-01-18,Salary,5000,income",
  "2026-01-19,Coffee,-120,food",
]

output = summarize_transactions(lines)
print(output)
# print(json.dumps(output,indent=4))
```
