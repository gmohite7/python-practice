```python
#local and global variable
a = 4

def temp():
    #local variable
    b = 3
    print(b)

print(a)

temp()
```

```python
# local and global same variable

a = 4

def temp():
    a = 2
    print(a)

temp()
print(a)
```

```python
a = 2

def temp():
    print(a)

temp()
print(a)
```

```python
#editing global variable in local namespace
a = 2

def temp():
    a = a + 2
    print(a)

temp()
print(a)
```

```python
#fixing the above error
a = 2

def temp():
    global a
    a = a + 2
    print(a)

temp()
print(a)
```

```python
# local and global.global created inside local

def temp():
    global a
    a = 4
    print(a)

temp()
print(a)
```
