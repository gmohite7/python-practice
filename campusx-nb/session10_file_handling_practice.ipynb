```python
#writing to a file

file = open('sample.txt','w')
file.write('Hello world')
file.close()
```

```python
#write multiline strings
file = open('sample.txt','w')
file.write('Hello world 1\n')
file.write('Hello world 2')
file.close()
```

```python
#using append mode
file = open('sample.txt','a')
file.write('\nSalman Khan\n')
file.write('Madhuri Dixit')
file.close()
```

```python
#write lines
lines = ['My name is Girish\n','I am 50 years old\n','I live in Mumbai with my wife and kids \n']
file = open('sample.txt','a')
file.writelines(lines)
file.close()
```

```python
# reading from file
file = open('sample.txt','r')
s1 = file.read()
s2 = file.read()
print(s1,s2)
file.close()
```

```python
# read line
file = open('sample.txt','r')
lines = file.readlines()
print(lines)
file.close()
```

```python
# reading entire file using readline
file = open('sample.txt','r')
while True:
    data = file.readline()
    if data == '':
        break
    else:
        print(data)
file.close()
```

```python
#using context manager

with open('sample.txt','a') as file:
    file.write('Shahrukh Khan')
```

```python
#serialization of a json file
L = [1,2,3,4]
import json

with open('sample.json','w') as file:
    json.dump(L,file)
```

```python
#serialize dictionary

dict = {
    'name':'Girish Mohite',
    'age':'50',
    'height':'5.11'
}
import json
with open('sample.json','w') as file:
    json.dump(dict,file,indent=4)
```
